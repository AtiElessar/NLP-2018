# Downloading the corpus and computing tokens, types and TTR

```python
from nltk.tokenize import word_tokenize
f = file("TomSawyer", 'r')
s = f.read()
s = s.replace("--", ' ')
#Converting the utf-8 encoding of " to "
s = s.replace("\xe2\x80\x9c", '"')
s = s.replace("\xe2\x80\x9d", '"')
l = list(s)
for i in l:
    if i in '";,[{(!?:)}]"`':
        l.remove(i)

s = ''.join(l)

def tokenise(st):

    l = word_tokenize(st)


    while '' in l:
        l.remove('')
    while '.' in l:
        l.remove('.')
    for i in range(len(l)):
            l[i] = l[i].strip("'")


    tokens = len(l)
    types = len(set(l))
    ttr = float(types)/tokens
    
    return tokens,types,ttr
    
    

tokens,types,ttr = tokenise(s)

print "Tokens = ", tokens
print "Types = ", types
print "TTR = ", ttr

```

# Zipf's Laws

```python
from collections import Counter
import matplotlib.pyplot as plt
l = s.split()
c = Counter(l)
freq = c.values()
freq = list(set(freq))
freq.sort(reverse = True)
plt.plot(freq)
plt.ylabel("Frequency")
plt.xlabel("Rank")
plt.show()
print "\t\tZipf's Law - 1"

```

As seen from the graph, Frequency $\propto \frac{1}{Rank}$

```python
freq_mean = {}
# The number of meanings have been taken from http://www.gutenberg.org/cache/epub/29765/pg29765.txt
meanings = {"time":16, "long":13,"see":11, "want" : 9, "find":7, "moment" : 6, "fool":5, "pocket":4, "whitewash":3,  "platform":2, 'imperishable':1}
for i in meanings:
    freq_mean[meanings[i]] = c[i]
    

lists = sorted(freq_mean.items())
mean, freq2 = zip(*lists)

plt.plot(mean,freq2,'o')
plt.ylabel("Frequency")
plt.xlabel("Number of meanings")
plt.show()
print "\t\tZipf's Law - 2"
```

Frequency $\propto$ Number of meanings $^2$

This Zipf's law does not work on function words and is not very accurate even for content words. 
Many content words, while having same number of meanings, have different frequencies. 
Similarly, many words with same frequency have different number of meanings.

```python
length = {}

for i in c:
    try:
        length[len(i)].append(c[i])
    except:
        length[len(i)] = [c[i]]
        
for i in length:
    length[i] = sum(length[i])/len(length[i])
    
length_rank = []

li = length.keys()
li.sort()

for i in li:
    length_rank.append(length[i])

    
    
plt.plot(length_rank)
plt.ylabel("Avg. Frequency")
plt.xlabel("Length")
plt.show()
print "\t\tZipf's Law - 3"
```

As seen from the graph, Frequency $\propto \frac{1}{Length \, of \, word}$

# Heap's Law

```python
lines = s.split('\n')
heap_tokens = []
heap_types = []
for z in xrange(100, 8844, 100):
    stemp = '\n'.join(lines[:z+1])
    ttokens,ttypes,tttr = tokenise(stemp)
    heap_tokens.append(ttokens)
    heap_types.append(ttypes)
    
    
heap_tokens.append(tokens)
heap_types.append(types)

plt.plot(heap_tokens,heap_types)
plt.ylabel("Vocabulary size V")
plt.xlabel("Tokens N")
plt.show()

    
```

## Fitting the curve

```python
from scipy.optimize import curve_fit

def curve(n,k,b):
     return k*(n**b)
    
popt, pcov = curve_fit(curve, heap_tokens, heap_types)
k = popt[0]
b = popt[1]
print "K = ",k,'\n', "Beta =",b

plt.figure(figsize=(20,15))
plt.plot(heap_tokens,heap_types, 'yo')
plt.plot(heap_tokens, curve(heap_tokens,k,b), 'r')
plt.ylabel("Vocabulary size V")
plt.xlabel("Tokens N")
plt.legend(["Experimental Value", "Fitted curve"])
plt.show()
```

V = kN$^\beta$
